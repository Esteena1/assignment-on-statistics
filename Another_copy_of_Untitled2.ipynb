**Case Study on Testing of hypothesis**


```python
df=pd.read_csv(r'/content/Sales_add.csv')
df
```

**2.a..**

 H_0 : mu_1 = mu_2 (There is no significant difference in sales before and after stepping into digital marketing)

H_1: mu_1 < mu_2 (there is an increase in sales after stepping into digital marketing)

```python
from scipy import stats

t_statistic,p=stats.ttest_rel(df['Sales_before_digital_add(in $)'], df['Sales_After_digital_add(in $)'])
print("t-statistic = ",round(t_statistic,2),"\n p_value = ",p)
alpha=0.05
if(p<alpha):
  print("Reject H_0, i.e., there is an increase in sales after stepping into digital marketing")
else:
  print("There is no significant difference in sales before and after stepping into digital marketing")
```

**2.b..**

```python
crosstab= pd.crosstab(df['Region'],df['Manager'])
print(crosstab)
print(stats.chi2_contingency(crosstab))

```

Here p_value is greater than 0.05, so we fail to  rejct the null hypothesis i.e., there is no relationship between the features 'Manager' and ' Region'
